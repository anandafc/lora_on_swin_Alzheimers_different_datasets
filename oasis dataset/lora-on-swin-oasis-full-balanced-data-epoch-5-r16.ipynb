```python
import torch
from torch.utils.data import DataLoader, Dataset, WeightedRandomSampler
import torchvision.transforms as transforms
from PIL import Image
from transformers import AutoImageProcessor, AutoModelForImageClassification
from peft import LoraConfig, get_peft_model
from accelerate import Accelerator
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report, recall_score, precision_score
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
import warnings
import albumentations as A
from albumentations.pytorch import ToTensorV2
from pathlib import Path
warnings.filterwarnings('ignore')

# Check if GPU is available
print("GPU available:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))
```

```python
# Find the actual dataset structure
dataset_path = '/kaggle/input/imagesoasis'

# Recursively search for image files to understand the structure
image_paths = []
for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_paths.append(os.path.join(root, file))

print(f"Found {len(image_paths)} images total")

# Discover the actual folder structure
folder_structure = {}
for img_path in image_paths[:10]:  # Check first 10 images to understand structure
    relative_path = os.path.relpath(img_path, dataset_path)
    folders = relative_path.split(os.sep)
    print(f"Image path: {relative_path}")
    
    # Track folder structure
    for i, folder in enumerate(folders[:-1]):  # Exclude filename
        if folder not in folder_structure:
            folder_structure[folder] = set()
        folder_structure[folder].add(folders[i+1] if i+1 < len(folders)-1 else "IMAGE")

print("\nFolder structure analysis:")
for folder, contents in folder_structure.items():
    print(f"{folder}: {contents}")

# Let's find where the actual class folders are
class_candidates = []
for root, dirs, files in os.walk(dataset_path):
    if any(f.lower().endswith(('.png', '.jpg', '.jpeg')) for f in files):
        # This folder contains images directly
        class_name = os.path.basename(root)
        num_images = len([f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
        if num_images > 0:
            class_candidates.append((class_name, root, num_images))

print("\nPotential class folders:")
for class_name, path, count in class_candidates:
    print(f"{class_name}: {count} images at {path}")

# Use the discovered class folders
if class_candidates:
    class_names = [name for name, path, count in class_candidates]
    dataset_path = os.path.dirname(class_candidates[0][1])  # Get parent directory
    print(f"\nUsing classes: {class_names}")
    print(f"Dataset path: {dataset_path}")
else:
    # Fallback: look for any folders that might be classes
    print("No class folders found. Listing all directories:")
    all_items = os.listdir(dataset_path)
    for item in all_items:
        item_path = os.path.join(dataset_path, item)
        if os.path.isdir(item_path):
            print(f"Directory: {item}")
            # Check if it contains subdirectories that might be classes
            subitems = os.listdir(item_path)
            for subitem in subitems:
                subitem_path = os.path.join(item_path, subitem)
                if os.path.isdir(subitem_path):
                    images = [f for f in os.listdir(subitem_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
                    if images:
                        print(f"  -> {subitem}: {len(images)} images")
                        class_names = [subitem]
                        dataset_path = item_path
                        break

class_to_idx = {name: idx for idx, name in enumerate(class_names)}
print(f"Final class mapping: {class_to_idx}")
```

```python
# Offline augmentation for 'Moderate Dementia' to boost sample size
minority_path = os.path.join(dataset_path, 'Moderate Dementia')
augmented_path = '/kaggle/working/augmented_moderate'
Path(augmented_path).mkdir(exist_ok=True)

# Define augmentation pipeline for offline augmentation (RGB-compatible)
offline_transform = A.Compose([
    A.Resize(224, 224),  # Ensure consistent size
    A.RandomRotate90(p=0.5),
    A.HorizontalFlip(p=0.5),  # Safe for MRIs
    A.VerticalFlip(p=0.5),    # Safe for MRIs
    A.GaussNoise(p=0.2),      # Adds noise, suitable for RGB
    A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.3),  # RGB-compatible
])

# Generate augmented images for 'Moderate Dementia'
for img_file in os.listdir(minority_path):
    if img_file.endswith(('.jpg', '.png', '.jpeg')):
        img = Image.open(os.path.join(minority_path, img_file)).convert('RGB')  # Load as RGB
        img_np = np.array(img)
        for i in range(20):  # Create 20 augmented versions per image
            augmented = offline_transform(image=img_np)['image']  # Correct syntax
            Image.fromarray(augmented).save(os.path.join(augmented_path, f'aug_{i}_{img_file}'))

# Copy original 'Moderate Dementia' images to augmented folder (resized)
for img_file in os.listdir(minority_path):
    if img_file.endswith(('.jpg', '.png', '.jpeg')):
        img = Image.open(os.path.join(minority_path, img_file)).convert('RGB')  # Load as RGB
        img_resized = img.resize((224, 224))  # Ensure consistent size
        img_resized.save(os.path.join(augmented_path, img_file))

# Update dataset path to include augmented images
class_names = ['Non Demented', 'Very mild Dementia', 'Moderate Dementia', 'Mild Dementia']
dataset_path = '/kaggle/input/imagesoasis/Data'
class_to_idx = {'Non Demented': 0, 'Very mild Dementia': 1, 'Moderate Dementia': 2, 'Mild Dementia': 3}
# Update path for Moderate Dementia to augmented folder
for i, (name, path, count) in enumerate(class_candidates):
    if name == 'Moderate Dementia':
        class_candidates[i] = (name, augmented_path, len(os.listdir(augmented_path)))

print("\nUpdated class counts after augmentation:")
for class_name, path, count in class_candidates:
    print(f"{class_name}: {count} images at {path}")
```

```python
# Define custom dataset class with class-specific augmentation
class AlzheimerDataset(Dataset):
    def __init__(self, root_dir, class_names, class_to_idx, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.class_names = class_names
        self.class_to_idx = class_to_idx
        self.images = []
        self.labels = []
        
        # Define strong augmentation for minority classes (RGB-compatible)
        self.minority_classes = [class_to_idx['Moderate Dementia'], class_to_idx['Mild Dementia'], class_to_idx['Very mild Dementia']]
        self.strong_transform = A.Compose([
            A.Resize(224, 224),  # Ensure consistent size
            A.RandomRotate90(p=0.5),
            A.HorizontalFlip(p=0.5),  # Safe for MRIs
            A.VerticalFlip(p=0.5),    # Safe for MRIs
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=30, p=0.5),
            A.GaussNoise(p=0.2),
            A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.3),  # RGB-compatible
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),  # 3-channel normalization
            ToTensorV2()
        ])
        
        # Load images from each class folder
        for class_name in class_names:
            class_path = os.path.join(root_dir, class_name) if class_name != 'Moderate Dementia' else augmented_path
            if os.path.exists(class_path):
                image_files = [f for f in os.listdir(class_path) 
                              if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
                
                for img_name in image_files:
                    self.images.append(os.path.join(class_path, img_name))
                    self.labels.append(class_to_idx[class_name])
                
                print(f"Loaded {len(image_files)} images from {class_name}")
            else:
                print(f"Warning: Class folder {class_name} not found!")
        
        if len(self.images) == 0:
            raise ValueError("No images found! Check dataset structure.")
        
        print(f"\nTotal: {len(self.images)} images from {len(class_names)} classes")
    
    def __len__(self):
        return len(self.images)
    
    def __getitem__(self, idx):
        img_path = self.images[idx]
        try:
            image = Image.open(img_path).convert('RGB')  # Load as RGB
            label = self.labels[idx]
            image_np = np.array(image)
            
            if label in self.minority_classes:
                # Apply strong augmentation for minorities
                augmented = self.strong_transform(image=image_np)['image']  # Correct syntax
            else:
                # Apply standard transform for majority
                if self.transform:
                    augmented = self.transform(image=image_np)['image']  # Correct syntax
                else:
                    # Fallback: resize and convert to tensor
                    augmented = A.Compose([
                        A.Resize(224, 224),
                        A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
                        ToTensorV2()
                    ])(image=image_np)['image']
            
            return augmented, label
        except Exception as e:
            print(f"Error loading {img_path}: {e}")
            return torch.zeros(3, 224, 224), 0  # Return 3-channel placeholder
```

```python
# Data transformations for majority class (Non Demented, RGB-compatible)
train_transform = A.Compose([
    A.Resize(224, 224),  # Ensure consistent size
    A.RandomCrop(224, 224, p=1.0),  # Already 224x224, but kept for consistency
    A.HorizontalFlip(p=0.5),
    A.RandomRotate90(p=0.5),
    A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.3),  # RGB-compatible
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),  # 3-channel normalization
    ToTensorV2()
])

test_transform = A.Compose([
    A.Resize(224, 224),  # Ensure consistent size
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),  # 3-channel normalization
    ToTensorV2()
])

# Load dataset
full_dataset = AlzheimerDataset(dataset_path, class_names, class_to_idx, transform=train_transform)

# Compute class weights for loss function
class_counts = np.bincount(full_dataset.labels)
class_weights = compute_class_weight('balanced', classes=np.unique(full_dataset.labels), y=full_dataset.labels)
class_weights = torch.tensor(class_weights, dtype=torch.float)

# Split dataset
train_size = int(0.8 * len(full_dataset))
test_size = len(full_dataset) - train_size
train_dataset, test_dataset = torch.utils.data.random_split(
    full_dataset, [train_size, test_size], 
    generator=torch.Generator().manual_seed(42)
)

# Apply test transform to test dataset
test_dataset.dataset.transform = test_transform

# Compute sample weights for weighted random sampling
sample_weights = [1.0 / class_counts[full_dataset.labels[idx]] for idx in train_dataset.indices]
sample_weights = torch.tensor(sample_weights, dtype=torch.double)
sampler = WeightedRandomSampler(sample_weights, len(sample_weights), replacement=True)

# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=16, sampler=sampler, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False, num_workers=2)

print(f"Train samples: {len(train_dataset)}, Test samples: {len(test_dataset)}")
```

```python
# Modify Swin Transformer to accept 1-channel (grayscale) inputs
def modify_swin_for_grayscale(model):
    try:
        # Original conv layer: expects 3 channels (RGB)
        original_conv = model.swin.embeddings.patch_embeddings.projection
        print(f"Original conv layer: in_channels={original_conv.in_channels}, out_channels={original_conv.out_channels}")

        # New conv layer: 1 input channel, same output channels, kernel size, stride, padding
        new_conv = torch.nn.Conv2d(
            in_channels=1,
            out_channels=original_conv.out_channels,
            kernel_size=original_conv.kernel_size,
            stride=original_conv.stride,
            padding=original_conv.padding,
            bias=original_conv.bias is not None
        )
        
        # Initialize new weights by averaging the pre-trained RGB weights across channels
        with torch.no_grad():
            new_conv.weight.copy_(original_conv.weight.mean(dim=1, keepdim=True))  # Average across input channels
            if new_conv.bias is not None:
                new_conv.bias.copy_(original_conv.bias)
        
        # Replace the original conv layer
        model.swin.embeddings.patch_embeddings.projection = new_conv
        
        # Verify the modification
        print(f"Modified conv layer: in_channels={model.swin.embeddings.patch_embeddings.projection.in_channels}, out_channels={model.swin.embeddings.patch_embeddings.projection.out_channels}")
        if model.swin.embeddings.patch_embeddings.projection.in_channels != 1:
            raise ValueError("Failed to modify Swin Transformer to accept 1-channel input!")
        
        return model
    except Exception as e:
        raise RuntimeError(f"Error modifying Swin Transformer for grayscale: {e}")
```

```python
# Load pre-trained Swin Transformer
model_name = "microsoft/swin-tiny-patch4-window7-224"
processor = AutoImageProcessor.from_pretrained(model_name)
model = AutoModelForImageClassification.from_pretrained(
    model_name,
    num_labels=len(class_names),
    ignore_mismatched_sizes=True
)

print(f"Swin Transformer loaded for {len(class_names)} classes, expecting RGB input!")
```

```python
# Define LoRA configuration
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["query", "key", "value", "dense"],
    lora_dropout=0.1,
    bias="none",
)

# Apply LoRA to model
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
```

```python
# Initialize accelerator
accelerator = Accelerator()

# Set up optimizer and weighted loss function
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=0.01)
criterion = torch.nn.CrossEntropyLoss(weight=class_weights.to(accelerator.device))

# Prepare with accelerator
model, optimizer, train_loader, test_loader = accelerator.prepare(
    model, optimizer, train_loader, test_loader
)

print("Training setup complete!")
```

```python
# Training loop
model.train()
for epoch in range(5):
    total_loss = 0
    correct = 0
    total = 0
    
    for batch_idx, (inputs, labels) in enumerate(train_loader):
        outputs = model(inputs)
        loss = criterion(outputs.logits, labels)
        
        accelerator.backward(loss)
        optimizer.step()
        optimizer.zero_grad()
        
        total_loss += loss.item()
        
        # Calculate accuracy
        preds = torch.argmax(outputs.logits, dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)
        
        if batch_idx % 50 == 0:
            batch_acc = (preds == labels).float().mean().item()
            print(f'Epoch {epoch+1}/5 | Batch {batch_idx}/{len(train_loader)} | Loss: {loss.item():.4f} | Acc: {batch_acc:.4f}')
    
    epoch_acc = correct / total
    print(f'Epoch {epoch+1} completed | Avg Loss: {total_loss/len(train_loader):.4f} | Train Acc: {epoch_acc:.4f}')

print("Training completed!")
```

```python
# Evaluation
model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for inputs, labels in test_loader:
        outputs = model(inputs)
        preds = torch.argmax(outputs.logits, dim=1)
        
        all_preds.extend(accelerator.gather(preds).cpu().numpy())
        all_labels.extend(accelerator.gather(labels).cpu().numpy())

print("Evaluation completed!")
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report, recall_score, precision_score
import matplotlib.pyplot as plt
import numpy as np

# Calculate performance metrics
accuracy = accuracy_score(all_labels, all_preds)
f1 = f1_score(all_labels, all_preds, average='weighted')
recall = recall_score(all_labels, all_preds, average='weighted')
precision = precision_score(all_labels, all_preds, average='weighted')
cm = confusion_matrix(all_labels, all_preds)

print(f"Accuracy: {accuracy:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"Recall: {recall:.4f}")
print(f"Precision: {precision:.4f}")
print("\nClassification Report:")
print(classification_report(all_labels, all_preds, target_names=class_names))

# Plot confusion matrix
plt.figure(figsize=(10, 8))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix - Alzheimer\'s Disease Classification')
plt.colorbar()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=45)
plt.yticks(tick_marks, class_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

# Add text annotations
for i in range(len(class_names)):
    for j in range(len(class_names)):
        plt.text(j, i, str(cm[i, j]),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > cm.max()/2 else "black")

plt.tight_layout()
plt.show()

# Plot accuracy by class
class_accuracies = cm.diagonal() / cm.sum(axis=1)
plt.figure(figsize=(10, 6))
plt.bar(class_names, class_accuracies)
plt.title('Accuracy by Alzheimer\'s Disease Class')
plt.ylabel('Accuracy')
plt.xticks(rotation=45)
plt.ylim(0, 1)
for i, v in enumerate(class_accuracies):
    plt.text(i, v + 0.01, f'{v:.3f}', ha='center')
plt.tight_layout()
plt.show()
```
